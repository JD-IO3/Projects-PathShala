<h2>EXPLORATORY DATA ANALYSIS</h2>

<h4>WHAT IS THE OVERALL SALES TREND?</h4>

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#Getting month year from the dataset
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
# df['month_year']
```

```python
df_trend = df.groupby('month_year').agg({'sales': 'sum'}).reset_index()
df_trend
```

```python
#Plotting the graph
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'], df_trend['sales'], color='#b80045')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

<h4>WHICH ARE THE TOP 10 PRODUCTS BY SALES?</h4>

```python
product_sales = pd.DataFrame(df.groupby('product_name').agg({'sales': 'sum'}))
```

```python
product_sales = product_sales.sort_values('sales', ascending=False)
```

```python
#TOP 10 PRODUCTS BY SALES
product_sales[:10]
```

<h4>WHICH ARE THE MOST SELLING PRODUCTS</h4>

```python
most_sell_prod = pd.DataFrame(df.groupby('product_name').agg({'quantity': 'sum'}))
```

```python
most_sell_prod = most_sell_prod.sort_values('quantity', ascending=False)
```

```python
#TOP 10 MOST SELLING PRODUCTS
most_sell_prod[:10]
```

<h4>WHAT IS THE MOST PREFERRED SHIP MODEL?</h4> 

```python
plt.figure(figsize=(10,8.5))
sns.countplot(df['ship_mode'])
```

<h4>WHICH ARE THE MOST PROFITABLE CATEGORY AND SUB-CATEGORY?</h4>

```python
cat_subcat_profit = pd.DataFrame(df.groupby(['category', 'sub_category']).agg({'profit': 'sum'}))
```

```python
cat_subcat_profit.sort_values(['category', 'profit'], ascending=False)
```
